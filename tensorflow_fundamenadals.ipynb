Aggregation tensors

```python
#get abs values
D=tf.constant([-7,10])
```

```python
tf.abs(D)
```

```python
E=tf.constant(np.random.randint(0,100,size=100))
```

```python
#find minimum
tf.reduce_min(D)
```

```python
tf.reduce_max(D)
```

```python
import tensorflow_probability as tfp
```

```python
tfp.stats.variance(D)
```

```python
tf.math.reduce_std(tf.cast(D,dtype=tf.float16))
```

```python
tf.math.reduce_variance(tf.cast(D,dtype=tf.float16))
```
